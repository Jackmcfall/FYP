# Comparing csv files 

```python
df1 = pd.read_csv("SerieA_DB.csv")
df2 = pd.read_csv("SerieA_DB_2.0.csv")
```

```python
df1[['Home', 'Away']] = df1[['Home', 'Away']].astype("str").astype("str")
df2[['Home', 'Away']] = df2[['Home', 'Away']].astype("str").astype("str")
```

```python
df1["Score"]==df2["Score"]
```

```python
assert df1.shape[0] == df2.shape[0], "Dataframes have different number of rows"
```

```python
merged_df = df1.merge(df2, on=['Date', 'Home', 'Away', 'Score'])
merged_df
```

```python
matching_columns = ['Score']

for i in range(df1.shape[0]):
    row1 = df1.iloc[i][matching_columns]
    row2 = df2.iloc[i][matching_columns]
    if not (row1 == row2).all():
        print(f"Row {i} does not match: {row1} vs {row2}")
```

```python
matching_columns = ['Home', 'Date', 'Away', 'Score']

for i in range(df1.shape[0]):
    row1 = df1.iloc[i][matching_columns]
    row2 = df2.iloc[i][matching_columns]
    print((row1 == row2).all())
```
